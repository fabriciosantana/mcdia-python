```python
import names
def create_group_of_elements(df: pd.DataFrame , group_size: int):

    df = pd.DataFrame(df).sample(frac=1, ignore_index=True)

    df.index = [f'Grupo {i // group_size}' for i in range(df.size)]
    
    return df

test_data = [names.get_full_name() for n in range(18)]

create_group_of_elements(test_data, 2)
```
